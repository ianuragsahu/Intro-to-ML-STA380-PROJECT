Question: 
As Reuters is one of the biggest and most reputed news agencies, it is important that the news they report does not have an underlying bias or agenda. The articles published by their authors should be as objective as possible and free from strong emotions. Hence the question: How neutral are the articles published by Reuters really?

Approach: 
This is a question that can be answerd by running a sentiment analysis on the corpus available. We use the VADER Lexicon to do so because it is popular for online textual works and is empirically validated by multiple independent human judges. We also use SentimentIntensityAnalyzer from the Natural Language Toolkit which works for the VADER Lexicon. As we are not building a predictive model of any sort, we combine both the training and testing datasets for a complete analysis. We then find out the average neutrality of all the authors and then plot the neutrality scores for each author.

```python
import os
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt

#VADER Lexicon
nltk.download('vader_lexicon')

#Paths
train_data_path = '/Users/raghavvaidya/Downloads/ReutersC50/C50train'
test_data_path = '/Users/raghavvaidya/Downloads/ReutersC50/C50test'

#Initializing Sentiment Analyzer and Neutrality Dictionary
sia = SentimentIntensityAnalyzer()
author_neutrality_scores = {}

all_data_paths = [train_data_path, test_data_path]

#Looping through both the directories
for data_path in all_data_paths:
    for author_dir in os.listdir(data_path):
        author_path = os.path.join(data_path, author_dir)
        if os.path.isdir(author_path):
            article_neutrality_scores = []
            
            #Looping through articles in a certain directory
            for article_file in os.listdir(author_path):
                article_path = os.path.join(author_path, article_file)
                with open(article_path, 'r', encoding='utf-8') as f:
                    article_text = f.read()
                    sentiment_score = sia.polarity_scores(article_text)
                    article_neutrality_scores.append(sentiment_score['neu'])
            
            #Calculating the average neutrality score for the author
            avg_neutrality_score = sum(article_neutrality_scores) / len(article_neutrality_scores)
            
            #Storing the average neutrality score for the author
            if author_dir in author_neutrality_scores:
                author_neutrality_scores[author_dir].append(avg_neutrality_score)
            else:
                author_neutrality_scores[author_dir] = [avg_neutrality_score]

#Printing the average neutrality scores for each author
tot_score = 0
for author, avg_score in author_neutrality_scores.items():
    tot_score = tot_score + ((avg_score[0] + avg_score[1])/2)
print('Average neutrality score for all authors:' + str(tot_score/len(author_neutrality_scores)))

#Extracting author names and average neutrality scores and sorting
sorted_authors = sorted(author_neutrality_scores.keys(), key=lambda x: sum(author_neutrality_scores[x]) / len(author_neutrality_scores[x]), reverse=True)
sorted_neutrality_scores = [sum(author_neutrality_scores[author]) / len(author_neutrality_scores[author]) for author in sorted_authors]

#Plotting
plt.figure(figsize=(10, 6))
plt.bar(sorted_authors, sorted_neutrality_scores, color='blue')
plt.xlabel('Authors')
plt.ylabel('Overall Average Neutrality Score')
plt.title('Distribution of Overall Average Neutrality Scores Among Authors')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Results: The average neutrality score for all the authors comes out around 0.8590706 which is 85.9%. As complete neutrality is expressed by a score of 1, the authors are really neutral. As seen in the graph the most neutral author is Jim Gilchrist and least neutral is Karl Penhaul.

Conclusion: All of the authors in the dataset are very neautral in their writing style as almost all of them achieve a neutrality score above 0.8. Hence we can confidently say by this analysis that the authors at Reuters approach the news very objectively.
